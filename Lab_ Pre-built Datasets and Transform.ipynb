```
import torch
import matplotlib.pyplot as plt
import numpy as np

torch.manual_seed(0)
```

```
def show_data(data_sample, shape=(28,28)):
  plt.imshow(data_sample[0].numpy().reshape(shape),cmap='gray')
  plt.title(' y = ' + str(data_sample[1]))
```

```
import torchvision.transforms as transforms
import torchvision.datasets as dsets
```

```
dataset = dsets.MNIST(
    root='./data',
    train = False,
    download = True,
    transform = transforms.ToTensor()
    )
```

```
print("Type of first element: ", type(dataset[0]))
print("The length of tuple: ",len(dataset[0]))
print("The shape of first element in tuple: ",dataset[0][0].shape)
print("Type of first element in tuple: ",type(dataset[0][0]))

print("The second element in the tuple: ",dataset[0][1])
print("Type of second element in tuple: ",type(dataset[0][1]))
#Issue as y element is not tensor
```

show_data(dataset[1])

```
croptensor_data_transform = transforms.Compose([transforms.CenterCrop(20),transforms.ToTensor()])
dataset = dsets.MNIST(root='./data',train=False,download=True,transform=croptensor_data_transform)
print("The shape of first element in first tuple: ",dataset[0][0].shape)
```

```
show_data(dataset[0],shape=(20,20))
```

```
fliptensor_data_transform = transforms.Compose([transforms.RandomHorizontalFlip(p=1), transforms.ToTensor()])
dataset = dsets.MNIST(root='./data',train=False,download=True,transform=fliptensor_data_transform)
show_data(dataset[0])
```

```
flip_horizontal_vertical = transforms.Compose([transforms.RandomVerticalFlip(),transforms.RandomHorizontalFlip(),transforms.ToTensor()])
dataset = dsets.MNIST(root='./data',train=False,download=True,transform=flip_horizontal_vertical)
show_data(dataset[0])
```
