```
import torch
from torch.utils.data import Dataset
torch.manual_seed(1)
```

```
class toy_set(Dataset):
  def __init__(self,length=100,transform=None):
    self.len = length
    self.x = 2 * torch.ones(length,2)
    self.y = torch.ones(length,1)
    self.transform = transform
  

  def __getitem__(self,index):
    sample = self.x[index], self.y[index]
    if self.transform:
      sample = self.transform(sample)
    return sample

  def __len__(self):
    return self.len
```

```
our_dataset = toy_set()
print("Our toy set object :",our_dataset)
print("Value on index 0 :",our_dataset[0])
print("length :",len(our_dataset))
```

```
class add_mult(object):
  def __init__(self, addx=1,muly = 2):
    self.addx = addx
    self.muly = muly
  
  def __call__(self,sample):
    x = sample[0]
    y = sample[1]

    x = x + self.addx
    y = y * self.muly

    sample = x, y
    return sample
```

```
a_m = add_mult()
data_set = toy_set()
```

```
print(data_set[0])
x_,y_ = a_m(data_set[0])
print(x_,y_)
```

```
cust_data_set = toy_set(transform = a_m)
print(cust_data_set[0])
```

```
from torchvision import transforms
```

```
class mult(object):
  def __init__(self,mult=100):
    self.mult = mult
  
  def __call__(self,sample):
    x = sample[0]
    y = sample[1]

    x = x * self.mult
    y = y * self.mult

    sample = x, y
    return sample
```

```
data_transform = transforms.Compose([add_mult(),mult()])
print("Data_transform :",data_transform)
```

```
composite_data_set = toy_set(transform=data_transform)
print(composite_data_set[0])
```
