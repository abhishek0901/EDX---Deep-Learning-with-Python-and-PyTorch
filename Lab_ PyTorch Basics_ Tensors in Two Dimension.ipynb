```
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
twoD_list = [[1,2,3],[4,5,6],[7,8,9]]
twoD_tensor = torch.tensor(twoD_list)
print("2Dim Tensor: ",twoD_tensor)
```

```
print("Dimensions :",twoD_tensor.ndimension())
print("Shape :",twoD_tensor.shape)
print("Size :",twoD_tensor.size())
```

```
to_np_array = twoD_tensor.numpy()
print(to_np_array)
print("=======================")
back_to_tensor = torch.from_numpy(to_np_array)
print(back_to_tensor)
```

```
df = pd.DataFrame({'a':[12,13,11],'b':[23,45,31]})
tensor_from_df = torch.from_numpy(df.values)
print(tensor_from_df)
```
